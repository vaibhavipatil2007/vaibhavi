```python
# scatter plot (to show connections, correction between 2 variable )

x=[1,2,3,4,5]
y=[5,4,7,10,8]
plt.scatter(x,y)
plt.title ("scatter plot")
```

```python
# heatmap (to show intercorrelation between numerical variable )

data = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 3, 2, 1],
    'C': [2, 3, 2, 3]
})
correlation_matrix = data.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
