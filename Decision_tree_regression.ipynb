```python
# Import Libraries 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# Importing and reading dataset 

ds=pd.read_csv("Position_Salaries.csv")
ds
```

```python
# splitting dataset into Independent and Dependent variable 

x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
x
```

```python
y
```

```python
# Fitting Decision Tree Regression model 

from sklearn.tree import DecisionTreeRegressor,plot_tree
reg = DecisionTreeRegressor(random_state=0)
reg.fit(x,y)
```

```python
# Prdict for my input 

myinput = pd.DataFrame(data=[[4]], columns=["Level"])
reg.predict(myinput)
```

```python
# Visualize Decision Tree 

plt.figure(figsize=(50,50))        # set figure
plot_tree(reg, feature_names=['Level'], filled=True)
plt.title("Decision Tree visualization" ,size=48)
plt.show()
```

```python
# For user input 

level = float(input("Enter Level: "))
myinput = pd.DataFrame(data=[[level]], columns=["Level"])
result = reg.predict(myinput)
print(f"Predicted Salary for Level {level}: {result[0]}")
```
