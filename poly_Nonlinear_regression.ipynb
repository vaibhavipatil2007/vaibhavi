```python
# Import Libraries 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# Importing and reading dataset 

ds = pd.read_csv("Position_Salaries.csv")
ds
```

```python
# splitting dataset into Indenpendent and Dependent 

x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
x
```

```python
y
```

```python
# Plot graph 

plt.scatter(x,y)
plt.show()
```

```python
# Model fitting (Polynimial regression model fitting)

from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=5)
x_poly = poly_reg.fit_transform(x)
#X_poly
reg=LinearRegression()
reg.fit(x_poly,y)
```

```python
# visualize 

plt.scatter(x,y)
plt.plot(x, reg.predict(poly_reg.fit_transform(x)),color='red')
plt.show()
```

```python
# predicting for my input 

myinput=pd.DataFrame(data=[[11]], columns=["Level"])
result=reg.predict(poly_reg.fit_transform(myinput))
result
```

```python
# Predict for user Input 

level = float(input("Enter Level: "))
myinput = pd.DataFrame(data=[[level]], columns=["Level"])
poly_input = poly_reg.transform(myinput)
result = reg.predict(poly_input)
print(f"Predicted result for Level {level}: {round(result[0,0])}")
```
