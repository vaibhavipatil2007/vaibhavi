```python
# Import Libraries 

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Importing and reading dataset 

ds = pd.read_csv("Position_Salaries.csv")
ds
```

```python
# splitting dataset into Indepedent and dependent variable 

x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
x
```

```python
y
```

```python
# Fitting random forest regression model 

from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(n_estimators=10, random_state=0)
reg.fit(x,y)
```

```python
# Predict for myinput

myinput = pd.DataFrame(data=[[4]], columns=["Level"])
reg.predict(myinput)
```

```python
# visualization

from sklearn.tree import plot_tree
plot_tree(reg.estimators_[0], filled=True)
plt.title("Random Forest Tree Regression" ,size=30)
plt.show()
```

```python
# Predict for user Input 

level = float(input("Enter the level: "))

myinput = pd.DataFrame(data=[[level]], columns=["Level"])
result = reg.predict(myinput)

print(f"Predicated  salary for Level {level} ₹{result[0]}")
```
