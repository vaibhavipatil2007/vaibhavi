```python
# handling missing values using mean and scikit-learn

from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan,strategy='mean')
ds[['Age','Salary']]=si.fit_transform(ds[['Age','Salary']])
```

```python
ds
```

```python
ds['Country'].unique
```

```python
# encoding categorical values(one-hot-encoding)

from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output = False)
encoded_data=ohe.fit_transform(ds[['Country']])
encoded_data
```

```python
# make column headers and make a dataframe for array

column_name=ohe.get_feature_names_out(['Country'])
encoded_df=pd.DataFrame(encoded_data,columns=column_name)
encoded_df
```

```python
# joining original dataset and new dataset

ds=pd.concat([ds, encoded_df],axis = 1)
ds
```

```python
# Droping catrgorical column

ds=ds.drop(columns=['Country'])
ds
```

```python
# label encoding

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds['Purchased']=le.fit_transform(ds['Purchased'])
```

```python
ds
```

```python
# feature scaling(reduce larger values to its smaller data values)

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
ds[['Age','Salary']]=sc.fit_transform(ds[['Age','Salary']])
```
